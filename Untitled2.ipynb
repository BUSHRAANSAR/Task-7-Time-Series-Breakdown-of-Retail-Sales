```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
from google.colab import files
uploaded = files.upload()

train = pd.read_csv("walmartrtrain.csv", parse_dates=["Date"])
stores = pd.read_csv("stores.csv")
features = pd.read_csv("features.csv", parse_dates=["Date"])

df = train.merge(features, on=["Date", "Store"], how="left")
df = df.merge(stores, on="Store", how="left")

df.set_index("Date", inplace=True)
monthly = df.resample("M").agg({
    "Weekly_Sales": "sum",
})

monthly["Sales_MA3"] = monthly["Weekly_Sales"].rolling(window=3).mean()
monthly["Sales_MA6"] = monthly["Weekly_Sales"].rolling(window=6).mean()

plt.figure(figsize=(12,6))
plt.plot(monthly.index, monthly["Weekly_Sales"], label="Total Sales")
plt.plot(monthly.index, monthly["Sales_MA3"], label="3-month MA")
plt.plot(monthly.index, monthly["Sales_MA6"], label="6-month MA")
plt.legend()
plt.title("Monthly Sales with Moving Averages")
plt.show()

from statsmodels.tsa.seasonal import seasonal_decompose

decomp = seasonal_decompose(monthly["Weekly_Sales"], model="additive")

fig = decomp.plot()

fig.set_size_inches(18, 10)

plt.suptitle("Seasonal Decomposition of Monthly Sales", fontsize=16)
plt.tight_layout()
plt.show()

df_reset = df.reset_index()
df_reset["Month"] = df_reset["Date"].dt.to_period("M")
grouped = df_reset.groupby(["Month", "Type"])["Weekly_Sales"].sum().unstack()
grouped.plot(figsize=(12,6))
plt.title("Sales by Store Type Over Time")
plt.ylabel("Monthly Sales")
plt.show()

model = ExponentialSmoothing(monthly["Weekly_Sales"], trend="add", seasonal="add", seasonal_periods=12)
fit = model.fit()
forecast = fit.forecast(12)

plt.figure(figsize=(12,6))
plt.plot(monthly.index, monthly["Weekly_Sales"], label="Observed")
plt.plot(forecast.index, forecast, label="Forecast", linestyle="--")
plt.legend()
plt.title("Forecast of Future Sales (Exponential Smoothing)")
plt.show()

```
